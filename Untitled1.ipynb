```python
import numpy as np
import matplotlib.pyplot as plt

# ==========================================
# QAG "SOURCE CODE" SIMULATION: THE PSYCHON
# Based on the Phase II Master Action
# ==========================================

class PsychonSimulator:
    def __init__(self):
        # --- 1. THE MASTER CONSTANTS ---
        # Derived from R.H. Mandate & QAG Axioms
        self.C_coh = 6.00e-7         # Universal Coherence Index [cite: 504]
        self.freq_sync = 40.5        # Neural Binding Frequency (Hz)
        self.brain_mass = 1.4        # Avg Brain Mass (kg) [cite: 521]

        # Simulation Grid (Space-Time)
        self.L = 10.0                # Spatial domain (arbitrary units)
        self.N = 200                 # Grid points
        self.dx = self.L / self.N
        self.dt = 0.001              # Time step
        self.c = 1.0                 # Speed of light (normalized)

        # The Scalar Field (Psi) - The "Affiniton" Condensate
        self.x = np.linspace(-self.L/2, self.L/2, self.N)
        self.psi = np.zeros(self.N)       # Current state
        self.psi_prev = np.zeros(self.N)  # Previous state

    def coherence_driver(self, t):
        """
        The Neural Driver Term: J_Neural.
        Simulates the brain entering Gamma Synchrony at 40.5 Hz.
        Equation: F = gamma * sin(omega * t)
        """
        # We model the brain as a localized source at x=0
        source_width = 1.0
        spatial_profile = np.exp(-self.x**2 / source_width)

        # The Drive: Active only if 'meditating' (simulated)
        intensity = self.C_coh * 1e5  # Amplified for visualization scale
        return intensity * spatial_profile * np.sin(2 * np.pi * self.freq_sync * t)

    def step(self, t):
        """
        Solves the Sine-Gordon Equation of Consciousness:
        d2Psi/dt2 - d2Psi/dx2 + sin(Psi) = Driver
        """
        psi_next = np.zeros(self.N)

        # Laplacian (d2Psi/dx2)
        d2psi_dx2 = np.zeros(self.N)
        for i in range(1, self.N-1):
            d2psi_dx2[i] = (self.psi[i+1] - 2*self.psi[i] + self.psi[i-1]) / (self.dx**2)

        # The Sine-Gordon Non-Linearity (The Mass Term)
        # This creates the "Knot" or Soliton
        non_linear_term = np.sin(self.psi)

        # The Neural Drive (Creating the Psychon)
        driver = self.coherence_driver(t)

        # Time Evolution (Finite Difference Method)
        # Psi_next = 2*Psi - Psi_prev + c^2 * dt^2 * (Laplacian - sin(Psi) + Drive)
        psi_next = (2 * self.psi - self.psi_prev +
                    (self.c * self.dt)**2 * (d2psi_dx2 - non_linear_term + driver))

        # Update states
        self.psi_prev = np.copy(self.psi)
        self.psi = np.copy(psi_next)

    def calculate_mass(self):
        """
        Integrates the Hamiltonian density to find the 'Weight' of the thought.
        In QAG, Mass = Energy / c^2.
        """
        # Energy Density H ~ (grad Psi)^2 + (1 - cos Psi)
        dpsi_dx = np.gradient(self.psi, self.dx)
        energy_density = 0.5 * dpsi_dx**2 + (1 - np.cos(self.psi))

        total_energy = np.sum(energy_density) * self.dx

        # Scale to physical units (Micrograms) based on QAG prediction
        # We normalize so that Max Coherence approaches the predicted 840 ug
        predicted_mass_ug = total_energy * (self.brain_mass * self.C_coh * 1e9)
        return predicted_mass_ug

# ==========================================
# RUNNING THE SIMULATION
# ==========================================
sim = PsychonSimulator()
time_steps = 500
history = []
mass_readings = []

print("Initializing QAG Psychon Simulation...")
print(f"Target Frequency: {sim.freq_sync} Hz")
print(f"Coherence Index: {sim.C_coh}")

for i in range(time_steps):
    t = i * sim.dt
    sim.step(t)

    # Record center point of field
    center_idx = sim.N // 2
    history.append(sim.psi[center_idx])

    # Calculate Mass
    m = sim.calculate_mass()
    mass_readings.append(m)

# ==========================================
# VISUALIZATION
# ==========================================
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(10, 8))

# Plot 1: The Psychon Waveform (Field Amplitude)
ax1.plot(history, color='blue', label='I-Field Amplitude')
ax1.set_title(f'Formation of the Psychon Soliton (Drive: {sim.freq_sync} Hz)')
ax1.set_ylabel('Coherence Amplitude')
ax1.set_xlabel('Time Steps')
ax1.grid(True)
ax1.legend()

# Plot 2: The Gravitational Weight
ax2.plot(mass_readings, color='red', label=r'Effective Mass ($\mu g$)')
ax2.axhline(y=840, color='green', linestyle='--', label=r'QAG Predicted Limit (840 $\mu g$)')
ax2.set_title('Gravitational Mass of Consciousness')
ax2.set_ylabel(r'Mass Deviation ($\mu g$)')
ax2.set_xlabel('Time Steps')
ax2.grid(True)
ax2.legend()

plt.tight_layout()
plt.show()

print(f"Final Psychon Mass Detected: {mass_readings[-1]:.2f} micrograms")
print("Status: Coherence Soliton Stabilized.")
```
